```python
def show_correlation_heatmap(data, sidesize = 35):
    
    # Set the width and height of the figure
    plt.figure(figsize = (sidesize, sidesize))
    
    correlations = data.select_dtypes(include = np.number).corr()

    plt.title('Correlation of parameters', fontsize = 30)
    hm = sns.heatmap(
        data = correlations, 
        annot = True,
        square = True, # make cells square
        cbar_kws = {'fraction' : 0.01}, # shrink colour bar
        cmap = 'OrRd', # use orange/red colour map
        linewidth = 1  # space between cells
    )
    
    hm.set_xticklabels(hm.get_xticklabels(), horizontalalignment = 'right', fontsize = 14, rotation = 45)
    hm.set_yticklabels(hm.get_yticklabels(), horizontalalignment = 'right', fontsize = 14)
```

```python
def observe_relationships_with_scatterplot(variable_x, variable_y, ax,color='green'):
    
    plt.xlabel(variable_x)
    plt.ylabel(variable_y)
    
    sns.scatterplot(data = df, x = variable_x, y = variable_y, ax = ax,color=color)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def observe_3d_relationships(x, y, z, ax):
    # Scatter plot
    sc = ax.scatter(df[x], df[y], df[z])

    # Setting axis labels
    ax.set_xlabel(x, labelpad=10)  # labelpad for extra spacing
    ax.set_ylabel(y, labelpad=10)
    ax.set_zlabel(z, labelpad=10)

    # Optional: Adding a title to each subplot for clarity
    ax.set_title(f'Relationship between {x}, {y}, and {z}')

# Example usage
fig = plt.figure(figsize=(14, 8))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
observe_3d_relationships('sweet_foods', 'oily_fried_foods', 'soft_drinks', ax1)

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
observe_3d_relationships('age', 'consult_specialist', 'hunger_during_the_day', ax2)

plt.show()
```

```python
def crosstab_barplot(feature_for_index, feature_for_columns, ax = None):
    
    crosstab_result = pd.crosstab(
        index = df[feature_for_index], 
        columns = df[feature_for_columns]
    )
    
    barplot = crosstab_result.plot.bar(stacked = True, rot = 0, ax = ax)
```

```python
fig = plt.figure(figsize = (16, 20))
fig.subplots_adjust(hspace = 0.4, wspace = 1)


ax = fig.add_subplot(5, 2, 1)
crosstab_barplot('order_go_out', 'diet_type_vegetarian', ax)


ax = fig.add_subplot(5, 2, 2)
crosstab_barplot('dairy_beverages', 'diet_type_eggetarian', ax)


ax = fig.add_subplot(5, 2, 3)
crosstab_barplot('skip_meals', 'restaurant_meals', ax)

ax = fig.add_subplot(5, 2, 4)
crosstab_barplot('alcoholic_beverages', 'restaurant_meals', ax)

ax = fig.add_subplot(5, 2, 5)
crosstab_barplot('fresh_fruit', 'main_meal_breakfast', ax)


ax = fig.add_subplot(5, 2, 6)
crosstab_barplot('cook_meals', 'main_meal_dinner', ax)


ax = fig.add_subplot(5, 2, 7)
crosstab_barplot('order_go_out', 'main_meal_dinner', ax)


ax = fig.add_subplot(5, 2, 8)
crosstab_barplot('skip_meals', 'main_meal_dinner', ax)

plt.show()
```
